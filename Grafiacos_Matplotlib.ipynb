```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.linspace(2,9,10)
x
```

```python
y= x**2
```

```python
y
```

```python
plt.plot(x,y,'red')
plt.title('Este es mi grafico')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'g')


plt.subplot(2,1,2)
plt.plot(y,x,'blue')
```

programacion orientada a objetos 

```python
figura = plt.figure()
grafico= figura.add_axes([0.1,0.1,1,1])
grafico.plot(x,y,'green')
plt.show()
```

```python
figura= plt.figure()
grafico1 = figura.add_axes([0.1,0.1,1,1])
grafico1.plot(x,y,'r')

grafico2 = figura.add_axes([0.2,0.5,0.3,0.3])
grafico2.plot(y,x,'blue')
plt.show()
```

Multigraficos(filas,columnas)

```python
figuras, graficos = plt.subplots(nrows=2,ncols=2)
```

```python
figuras, graficos = plt.subplots(nrows=2,ncols=2)
graficos[0][0].plot(x,y,'blue')
graficos[0][1].plot(x,y,'red')
graficos[1][0].plot(x,y,'y')
graficos[1][1].plot(x,y,'g')

graficos[0][0].set_title('Azul')
graficos[0][1].set_title('Rojo')
graficos[1][0].set_title('Amarillo')
graficos[1][1].set_title('Verde')
plt.show()
```

***tamao del grafico***

```python
figura= plt.figure(figsize=(8,4))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
plt.show()
```

Dos graficos en la misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2,label='Al cuadrado')
graficos.plot(x,x**3,label='Al cubo')
graficos.plot(x,x**4,label='Elevado a la cuarta')
graficos.legend(loc=0)
plt.show();
```

# Color del grfico

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
plt.plot(x,y,'red')
plt.show()
```

# Linea del grafico (grosor,transparencia y tipo)

```python
figura= plt.figure()
grafico= figura.add_axes([0,0,1,1])
grafico.plot(x,y,'red',linewidth=50,alpha=0.5,linestyle=':')
plt.show()
```

# Marcadores

```python
figura= plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'green', marker='o',markersize=5)
plt.show()
```
